```python
chemin_dossier = "../data/tweets"      # Chemin relatif

def lire_et_parser_json(chemin): # Fonction pour lire et parser le contenu d'un fichier JSON
    try:
        with open(chemin, 'r', encoding='utf-8') as fichier:
            return json.load(fichier)
    except json.JSONDecodeError as e:
        print(f"Erreur de d√©codage JSON : {e}")
        return None
    except FileNotFoundError as e:
        print(f"Fichier non trouv√© : {e}")
        return None
```

```python
def aff():

    for index, nom_fichier in enumerate(os.listdir(chemin_dossier)):
        if index < 10:
            chemin_complet = os.path.join(chemin_dossier, nom_fichier)
            data = lire_et_parser_json(chemin_complet)

            if data and 'text' in data and data['text'].strip():
                tt = TextBlob(data['text'])
                print(f"--- Fichier {index + 1}: {nom_fichier} ---")
                print(f"Texte: {tt}")
                print(f"Mots: {tt.words}")
                print(f"Phrases: {tt.sentences}")
                print(f"Tags POS: {tt.tags}")
                print(f"Sentiment: {tt.sentiment}")
                print(f"Polarit√©: {tt.polarity}")
                print(f"Subjectivit√©: {tt.subjectivity}")
                print(f"Phrases nominales: {tt.noun_phrases}")
                print(f"Texte corrig√©: {tt.correct()}")
                print(f"Texte tok√©niz√©: {TextBlob.tokenize(tt)}")
                print("\n")
        else:
            break


# Ex√©cuterla fonction dans un thread s√©par√©
thread = threading.Thread(target=aff)
thread.start()
```

```python
%%memit #info su rla m√©moire solliciter 
import os
import json
import pandas as pd
from textblob import TextBlob
import threading

def ma_fonction():
    colonnes = ['Fichier', 'ID', 'Texte', 'Mots', 'Phrases', 'Tags POS', 'Sentiment', 'Polarit√©', 'Subjectivit√©', 'Phrases nominales', 'Texte corrig√©', 'Texte tok√©niz√©']
    df = pd.DataFrame(columns=colonnes)
    data_list = []  # Liste pour accumuler les donn√©es

    for index, nom_fichier in enumerate(os.listdir(chemin_dossier)):
        if index < 500:  # limite a 500 fichier
            chemin_complet = os.path.join(chemin_dossier, nom_fichier)
            data = lire_et_parser_json(chemin_complet)

            if data is not None and 'text' in data and data['text'].strip():
                tt = TextBlob(data['text'])
                
                retweet_count = reply_count = like_count = quote_count = 0

                if 'metrics' in data and data['metrics']:
                    metric_key = list(data['metrics'].keys())[0]
                    retweet_count = data['metrics'][metric_key]['retweet_count']
                    reply_count = data['metrics'][metric_key]['reply_count']
                    like_count = data['metrics'][metric_key]['like_count']
                    quote_count = data['metrics'][metric_key]['quote_count']
                

                info = {
                    'Fichier': nom_fichier,
                    'ID': data['id'],
                    'Nb retweet':int(retweet_count),
                    'Nb like':int(like_count),
                    'Nb r√©ponses':int(reply_count),
                    'Nb citations': int(quote_count),
                    'Hashtags': data['hashtags'],
                    'Texte': str(tt),
                    'Mots': tt.words,
                    'Phrases': tt.sentences,
                    'Tags POS': tt.tags,
                    'Sentiment': tt.sentiment,
                    'Polarit√©': tt.polarity,
                    'Subjectivit√©': tt.subjectivity,
                    'Phrases nominales': tt.noun_phrases,
                    'Texte corrig√©': tt.correct(),
                    'Texte tok√©niz√©': tt.tokens
                }
                data_list.append(info)

    # Concat la liste de donn√©es avec le DataFrame
    df = pd.concat([df, pd.DataFrame(data_list)], ignore_index=True)
    return df

chemin_dossier = "../data/tweets" 
df_global = None

def wrapper_function():
    global df_global
    df_global = ma_fonction()

# Cr√©er et d√©marrer le thread
thread = threading.Thread(target=wrapper_function)
thread.start()
thread.join()  # Attendre la fin du thread

# df_global contient maintenant les donn√©es
```
