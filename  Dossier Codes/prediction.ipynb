```python
import numpy as np

# Créer des poids aléatoires pour chaque variable
weight_retweet = np.random.rand()
weight_like = np.random.rand()
weight_replies = np.random.rand()
weight_quotes = np.random.rand()
weight_positive = np.random.rand()
weight_negative = np.random.rand()

# Calculer le score avec des poids aléatoires
df['Score_random'] = (df['Nb retweet'] * weight_retweet) + (df['Nb like'] * weight_like) + (df['Nb réponses'] * weight_replies) + (df['Nb citations'] * weight_quotes) + (df['positive'] * weight_positive) + (df['negative'] * weight_negative)
```

```python
x_train = train_df[['Nb retweet','Nb like','Nb réponses', 'Nb citations','positive','negative']]
y_train_score = train_df['Normalized_Score']
y_train_score_random = train_df['Score_random']

x_test = test_df[['Nb retweet','Nb like','Nb réponses', 'Nb citations','positive','negative']]
y_test_score = test_df['Normalized_Score']
y_test_score_random = test_df['Score_random']
```

```python
def within_range_accuracy(y_true, y_pred, tolerance=0.1):
    """
    Calculer le pourcentage de prédictions qui se trouvent dans un intervalle défini par tolerance autour de y_true.
    
    Args:
    y_true: Les vraies valeurs.
    y_pred: Les valeurs prédites par le modèle.
    tolerance: La tolérance relative (par exemple, 0.1 pour 10%).

    Returns:
    Un Tensor représentant l'accuracy personnalisée.
    """
    tolerance = tf.constant(tolerance, dtype=tf.float32)
    lower_bound = y_true * (1 - tolerance)
    upper_bound = y_true * (1 + tolerance)
    within_range = tf.logical_and(y_pred >= lower_bound, y_pred <= upper_bound)
    accuracy = tf.reduce_mean(tf.cast(within_range, tf.float32))
    return accuracy
```

```python
import numpy as np
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping


input_shape = x_train.shape[1]

# Création du modèle
model = Sequential([
    Dense(256, activation='relu', input_shape=(input_shape,)),  
    Dense(128,activation='relu'),
    Dense(128,activation='relu'),
    Dense(1)  # Softplus pour garantir une sortie positive
])

# Compilation du modèle
model.compile(optimizer='adam',
              loss='mse',  # Mean Squared Error
              metrics=['mse','mae',within_range_accuracy])  # Mean Absolute Error pour évaluation

# Configuration du callback pour sauvegarder le meilleur modèle uniquement
checkpoint_cb = ModelCheckpoint("prediction/modele_cancer/modele_cancer.keras", save_best_only=True)

# Configuration du callback pour arrêter l'entraînement si aucune amélioration après 10 époques
early_stopping_cb = EarlyStopping(patience=10, restore_best_weights=True)



```

```python

model.summary()
#tf.keras.utils.plot_model(model, to_file='prediction/modele_cancer/model.png', show_shapes=True, show_layer_names=True, dpi=96)
```

```python
# Entraînement du modèle avec callbacks
history = model.fit(x_train, y_train_score, epochs=100,
                    validation_split=0.2,
                    callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
history_df = pd.DataFrame(history.history)
history_df.loc[0:, ['loss', 'val_loss']].plot()
print("Minimum Validation Loss: {:0.4f}".format(history_df['val_loss'].min()))
import matplotlib.pyplot as plt


# Sauvegarder le graphique dans un fichier
plt.savefig('prediction/modele_cancer/loss.png')
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
import numpy as np

# Supposons que vous avez déjà les variables x_train, y_train, x_test, y_test définies
# Si ce n'est pas le cas, vous devriez charger vos données ici et les préparer.

# Création de l'objet de régression linéaire
regr = linear_model.LinearRegression()

# Configuration de la validation croisée à K-folds sur l'ensemble d'entraînement
kf = KFold(n_splits=5, shuffle=True, random_state=1)

# Calcul des scores de validation croisée sur l'ensemble d'entraînement
scores = cross_val_score(regr, x_train, y_train_score, cv=kf, scoring='neg_mean_squared_error')
print("Erreurs quadratiques moyennes de chaque fold (négatives):", -scores)
print("Moyenne des erreurs quadratiques moyennes sur le train:", -scores.mean())

# Utilisation de cross_val_predict pour obtenir des prédictions de validation croisée sur l'ensemble d'entraînement
y_train_pred = cross_val_predict(regr, x_train, y_train_score, cv=kf)

# Calcul et affichage du coefficient de détermination R^2 sur l'ensemble d'entraînement
r2_train = r2_score(y_train_score, y_train_pred)
print("Coefficient of determination (R^2) on train set: %.2f" % r2_train)

# Entraînement du modèle sur l'ensemble d'entraînement complet
regr.fit(x_train, y_train_score)

# Faire des prédictions en utilisant l'ensemble de test
y_test_pred = regr.predict(x_test)

# Calcul et affichage de l'erreur quadratique moyenne sur l'ensemble de test
mse_test = mean_squared_error(y_test_score, y_test_pred)
print("Mean squared error on test set: %.2f" % mse_test)

# Calcul et affichage du coefficient de détermination R^2 sur l'ensemble de test
r2_test = r2_score(y_test_score, y_test_pred)
print("Coefficient of determination (R^2) on test set: %.2f" % r2_test)

# Affichage des coefficients du modèle
print("Coefficients: \n", regr.coef_)
```
